```python
observations = pd.crosstab(index=df['Class'], columns=df['Amount'],margins=True)
observations
```

```python
from scipy.stats import chi2_contingency

chi2, p, dof, ex = chi2_contingency(observations)

print("chi2 = ", chi2)
print("p = ", p)
print("degrees of freedom = ", dof)
print("expected = ", ex)
```

```python
Here p=1 which is >0.05 there is no significant relation between amount and class.So we dont have enough evidence to reject Null Hypothesis.
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold, train_test_split,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from imblearn.over_sampling import SMOTE

X = df.drop(columns=['Class']).reset_index(drop=True)
y = df['Class'].astype(int).reset_index(drop=True)

sm = SMOTE(k_neighbors=1, random_state=42)  # k=1 because you only have 2 positives
X_bal, y_bal = sm.fit_resample(X, y)
print(X_bal.shape, y_bal.value_counts().to_dict())

X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size=0.4)
```

```python
# Initialize Logistic Regression
logreg = LogisticRegression(max_iter=1000)
# Perform cross-validation and get F1 scores
f1_scores_logreg = cross_val_score(logreg, X_train, y_train, cv=5, scoring='f1_weighted')
acc_scores_logreg = cross_val_score(logreg, X_train, y_train, cv=5, scoring='accuracy')
# Print F1 scores across folds and mean F1 score
print("Logistic Regression - F1 Scores across folds:", f1_scores_logreg)
print("Logistic Regression - Mean F1 Score:", f1_scores_logreg.mean())
print("Logistic Regression - Accuracy Scores across folds:", acc_scores_logreg)
print("Logistic Regression - Mean Accuracy Score:", acc_scores_logreg.mean())

```

```python
# Initialize Support Vector Machine
svm = SVC()
# Perform cross-validation and get F1 scores
f1_scores_svm = cross_val_score(svm, X_train, y_train, cv=5, scoring='f1_weighted')
acc_scores_svm=cross_val_score(svm, X_train, y_train, cv=5, scoring='accuracy')
# Print F1 scores across folds and mean F1 score
print("Support Vector Machine - F1 Scores across folds:", f1_scores_svm)
print("Support Vector Machine - Mean F1 Score:", f1_scores_svm.mean())
print("Support Vector Machine - Accuracy Scores across folds:", acc_scores_svm)
print("Support Vector Machine - Mean Accuracy Score:", acc_scores_svm.mean())
```

```python
# Initialize Random Forest
rf = RandomForestClassifier(n_estimators=100, class_weight="balanced", random_state=42)
# Perform cross-validation and get F1 scores
f1_scores_rf = cross_val_score(rf, X_train, y_train, cv=5, scoring='f1')
acc_scores_rf=cross_val_score(rf, X_train, y_train, cv=5, scoring='accuracy')
# Print F1 scores across folds and mean F1 score
print("Random Forest - F1 Scores across folds:", f1_scores_rf)
print("Random Forest - Mean F1 Score:", f1_scores_rf.mean())
print("Random Forest- Accuracy Scores across folds:", acc_scores_rf)
print("Random Forest - Mean Accuracy Score:", acc_scores_rf.mean())
```
